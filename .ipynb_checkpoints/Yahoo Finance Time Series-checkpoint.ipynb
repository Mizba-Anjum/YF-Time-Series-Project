## Prepare Dataset

```python
# Find dataset path
import os.path
dataset_path = os.path.abspath(os.path.dirname("AAPL.csv"))
dataset_path = os.path.join(dataset_path, "AAPL.csv")
```

```python
# Read dataset
df = pd.read_csv(dataset_path)
```

```python
# Convert dates to datetime format
df['Date'] = pd.to_datetime(df.Date, utc=True)
# Set dates as index
df = df.set_index('Date')
# Fill missing dates with previous date
df = df.reindex(pd.date_range(df.index[0], df.index[-1], freq='D'), method='ffill')
# Convert to date (no time)
df.index = pd.to_datetime(df.index).date
# Drop dividends and stock splits columns
df.drop(columns = ["Dividends", "Stock Splits", "Volume"], inplace=True)
```

## Window

Class TS_Predictor
<br>
set start and end of training data (date)
window_size
label column name
predict the next day based on the past 'window_size' days
to predict x days into the future: make prediction for each of x days

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
class Data_Prepper():
    def __init__(self, label, window_size = 30, df=df, data_start='', data_end=''):
        self.df_to_use = df.copy() # dataset to use
        self.df_to_use = self.df_to_use[[label]]
        if data_start and data_end: # portion of dataset to use (default: use full dataset)
            self.set_data(data_start, data_end)
        self.window_size = window_size # predictions based on past 30 (or specified) days
    
    # set portion of dataset to use
    def set_data(self, data_start, data_end):
        ds = pd.to_datetime(data_start).date()
        de = pd.to_datetime(data_end).date()
        self.df_to_use = self.df_to_use[(self.df_to_use.index >= ds) & (self.df_to_use.index <= de)]
    
    # split and normalize training data
    def normalize_data(self):
        scaler = MinMaxScaler()
        norm_df = scaler.fit_transform(self.df_to_use)
        return norm_df
    
    # split into x (features) and y (label) lists
    def split_to_xy(self):
        df_as_np = self.normalize_data()
        x = []
        y = []
        for i in range(len(df_as_np) - self.window_size):
            row = [a for a in df_as_np[i:i+self.window_size]]
            x.append(row)
            y.append(df_as_np[i+self.window_size])
        return np.array(x), np.array(y)
    
    # split into train, valid, test data sets
    def split_to_train(self, x, y):
        # training data: 70%, validation data: 20%, test data: 10%
        x_train, x_valid, x_test = np.split(x, [int(.7*len(x)), int(.9*len(x))])
        y_train, y_valid, y_test = np.split(y, [int(.7*len(y)), int(.9*len(y))])
        return [x_train, x_valid, x_test], [y_train, y_valid, y_test]
    
    # return train, valid, test data sets
    def get_xy_data(self):
        x_data, y_data = self.split_to_train(*self.split_to_xy())
        return self.df_to_use, x_data, y_data
```

```python
dp_30 = Data_Prepper(label="Close", df=df, data_start='2015-01-01', data_end=df.index[-1].strftime("%Y-%m-%d"))
```

```python
trained_df, [x_train, x_valid, x_test], [y_train, y_valid, y_test] = dp_30.get_xy_data()
```

```python
import tensorflow as tf
from tensorflow import keras
import keras_tuner as kt
```

```python
def model_builder(hp):
    model = keras.Sequential()
    # input x
    model.add(keras.layers.InputLayer(input_shape=(x_train.shape[-2], x_train.shape[-1])))
    
    model.add(keras.layers.LSTM(64))

    # Tune the number of units in the first Dense layer
    # Choose an optimal value between 32-64
    hp_units = hp.Int('units', min_value=32, max_value=64, step=4)
    model.add(keras.layers.Dense(units=hp_units, activation='relu'))
    
    model.add(keras.layers.Dense(x_train.shape[-1], activation='linear'))

    # Tune the learning rate for the optimizer
    # Choose an optimal value from 0.01, 0.001, or 0.0001
    hp_learning_rate = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])

    model.compile(optimizer=keras.optimizers.Adam(learning_rate=hp_learning_rate),
                loss=keras.losses.MeanSquaredError(), metrics=[keras.metrics.RootMeanSquaredError()])

    return model
```

```python
tuner = kt.Hyperband(model_builder,
                     objective=kt.Objective(keras.metrics.RootMeanSquaredError().name, 'min'),
                     max_epochs=100,
                     factor=3,
                     directory='hypertuning_trials',)
```

```python
stop_early = tf.keras.callbacks.EarlyStopping(monitor='val_root_mean_squared_error', patience=5)
tuner.search(x_train, y_train, epochs=100, validation_data=(x_valid, y_valid))
```

```python
best_model = tuner.get_best_models()[0]
```

```python
best_model.summary()
```

```python
eval_result = best_model.evaluate(x_test, y_test)
print("[test loss, test accuracy]:", eval_result)
```

```python
# predict a day
def predict_day(day, update=False, df=df, model=best_model):
    # get previous 30 days
    day = pd.to_datetime(day).date()
    end_day = day - timedelta(days=1)
    start_day = df.iloc[df.index.get_loc(end_day) - 29].name
    pred_df = df[start_day:end_day].copy()
    pred_x_data = np.array([pred_df])
    
    # predict day
    y_pred = model.predict(pred_x_data)
    
    # if the day is in the dataframe, overwrite it with the prediction
    # if the day is not in the dataframe, add it
    if update:
        df.loc[day] = y_pred[0]
        
    return y_pred
```
